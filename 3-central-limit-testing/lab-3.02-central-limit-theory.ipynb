## Exercise 1 - Cholesterol 

A health survey between 1988 and 1994 indicated that the mean cholesterol value for women between the age of 20 and 29 was 183 mg/dl, with a standard deviation equal to 36. We now take a random sample of 81 women.

1. Plot the probability distribution of the sample mean $\overline{x}$.
2. What is the probability of the sample mean being $\overline{x} < 185$? (Answer ≈ 69,1%)
3. What is the probability of $175 < \overline{x} < 185$ (Answer ≈ 66,9%)
4. What is the probability of $\overline{x} > 190$ (Answer ≈ 4,0%)

```python
# Gegeven waarden
mu = 183       # populatiegemiddelde
sigma = 36     # standaarddeviatie
n = 81         # steekproefgrootte

# Bereken standaardfout (SE)
SE = sigma / np.sqrt(n)

# 1. Plot de kansverdeling van de steekproefgemiddelde x̄
x_values = np.linspace(mu - 4*SE, mu + 4*SE, 1000)
pdf_values = stats.norm.pdf(x_values, loc=mu, scale=SE)

plt.figure(figsize=(8,5))
plt.plot(x_values, pdf_values)
plt.title('Probability distribution of sample mean x̄')
plt.xlabel('Sample mean (x̄)')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# 2. P(x̄ < 185)
prob_xbar_less_185 = stats.norm.cdf(185, loc=mu, scale=SE)
print(f"P(x̄ < 185) = {prob_xbar_less_185:.4f} ({prob_xbar_less_185*100:.1f}%)")

# 3. P(175 < x̄ < 185)
prob_xbar_between_175_185 = stats.norm.cdf(185, loc=mu, scale=SE) - stats.norm.cdf(175, loc=mu, scale=SE)
print(f"P(175 < x̄ < 185) = {prob_xbar_between_175_185:.4f} ({prob_xbar_between_175_185*100:.1f}%)")

# 4. P(x̄ > 190)
prob_xbar_greater_190 = 1 - stats.norm.cdf(190, loc=mu, scale=SE)
print(f"P(x̄ > 190) = {prob_xbar_greater_190:.4f} ({prob_xbar_greater_190*100:.1f}%)")
```

## Exercise 2

A random sample of 64 is drawn from a population with unknown distribution. However, the expectation and standard deviation of the population are known: $\mu = 20$ and $\sigma = 16$.

1. Plot the probability distribution of the sample mean.
2. Calculate the z-score for $\overline{x_1} = 15.5$  
3. Calculate the z-score for $\overline{x_2} = 23$ 
4. Determine the probability that 16 < $\overline{x}$ < 22 (Answer ≈ 81.9%)

```python
# Gegeven waarden
mu = 20       # populatiegemiddelde
sigma = 16    # standaarddeviatie
n = 64        # steekproefgrootte

# Bereken standaardfout (SE)
SE = sigma / np.sqrt(n)

# 1. Plot de kansverdeling van de steekproefgemiddelde x̄
x_values = np.linspace(mu - 4*SE, mu + 4*SE, 1000)
pdf_values = stats.norm.pdf(x_values, loc=mu, scale=SE)

plt.figure(figsize=(8,5))
plt.plot(x_values, pdf_values)
plt.title('Probability distribution of sample mean x̄')
plt.xlabel('Sample mean (x̄)')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# 2. Z-score voor x̄₁ = 15.5
x1 = 15.5
z1 = (x1 - mu) / SE
print(f"Z-score voor x̄₁ = 15.5: {z1:.4f}")

# 3. Z-score voor x̄₂ = 23
x2 = 23
z2 = (x2 - mu) / SE
print(f"Z-score voor x̄₂ = 23: {z2:.4f}")

# 4. P(16 < x̄ < 22)
prob_between_16_22 = stats.norm.cdf(22, loc=mu, scale=SE) - stats.norm.cdf(16, loc=mu, scale=SE)
print(f"P(16 < x̄ < 22) = {prob_between_16_22:.4f} ({prob_between_16_22*100:.1f}%)")
```
